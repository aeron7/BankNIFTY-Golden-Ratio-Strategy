```python
range_high = banknifty_info['data'][0]['dayHigh']
range_low = banknifty_info['data'][0]['dayLow']
print(range_high)
print(range_low)

#Assuming that We shall be running the program in 9:25 i.e. 10 mins after the market opens
```

```python
#The Golden Ratio BankNIFTY Strategy
##Scrip = BANKNIFTY

## Golden Number = ((Previous Day High - Previous Day Low) + Opening Range of Today's First 10m))*61.8%

### Buy Above = (Previous Day Close + Golden Number)
### Sell Below = (Previous Day Close - Goldern Number)

#### Stop Loss at 0.5% Target at 2%
```

```python
opening_range = range_high-range_low
print(opening_range)
```

```python
golden_number = (float(previous_day_high)-float(previous_day_low)+opening_range)*.618
golden_number
```

```python
previous_day_close = banknifty_info['data'][0]['previousClose']
previous_day_close
```

```python
buy_above = previous_day_close+golden_number
sell_below = previous_day_close-golden_number

print(buy_above)
print(sell_below)
```
